```python
# Making the mental states df
mental_states_table = ['Hours per day', 'Frequency [Classical]', 'Frequency [Country]', 'Frequency [EDM]', 'Frequency [Folk]', 'Frequency [Gospel]', 'Frequency [Hip hop]', 'Frequency [Jazz]', 'Frequency [K pop]', 'Frequency [Latin]', 'Frequency [Lofi]', 'Frequency [Metal]', 'Frequency [Pop]', 'Frequency [R&B]', 'Frequency [Rap]', 'Frequency [Rock]', 'Frequency [Video game music]', 'Anxiety', 'Depression', 'Insomnia']
mental_states_df = mxmh_cleaned_df[mental_states_table]
mental_states_df.head()
```

```python
mental_states_df.head()
```

```python
# Renaming columns so there are no spaces
new_column_names = {'Hours per day': 'Hours_per_day', 
'Frequency [Classical]': 'Frequency_Classical', 
'Frequency [Country]': 'Frequency_Country', 
'Frequency [EDM]': 'Frequency_EDM', 
'Frequency [Folk]': 'Frequency_Folk', 
'Frequency [Gospel]': 'Frequency_Gospel', 
'Frequency [Hip hop]': 'Frequency_Hiphop', 
'Frequency [Jazz]': 'Frequency_Jazz', 
'Frequency [K pop]': 'Frequency_Kpop', 
'Frequency [Latin]': 'Frequency_Latin', 
'Frequency [Lofi]': 'Frequency_Lofi', 
'Frequency [Metal]': 'Frequency_Metal', 
'Frequency [Pop]': 'Frequency_Pop', 
'Frequency [R&B]': 'Frequency_RB', 
'Frequency [Rap]': 'Frequency_Rap', 
'Frequency [Rock]': 'Frequency_Rock', 
'Frequency [Video game music]': 'Frequency_Videogame', 
'Anxiety': 'Anxiety', 'Depression': 'Depression', 'Insomnia': 'Insomnia'}
mental_states_df.rename(columns=new_column_names, inplace=True)
```

```python
# Converting categorical values to numerics for ease during SQL set up
columns_to_convert = ['Frequency_Classical', 
'Frequency_Country', 
'Frequency_EDM', 
'Frequency_Folk', 
'Frequency_Gospel', 
'Frequency_Hiphop', 
'Frequency_Jazz', 
'Frequency_Kpop', 
'Frequency_Latin', 
'Frequency_Lofi', 
'Frequency_Metal', 
'Frequency_Pop', 
'Frequency_RB', 
'Frequency_Rap', 
'Frequency_Rock', 
'Frequency_Videogame']
conversion_dict = {'Never': '0', 'Rarely': '1', 'Sometimes': '2', 'Very frequently': '3'}

for column in columns_to_convert:
    mental_states_df[column] = mental_states_df[column].map(conversion_dict)
```

```python
mental_states_df = mental_states_df.drop(['Hours_per_day'], axis=1)
mental_states_df.head()
```

```python
import hvplot.pandas
import plotly.express as px
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
```

```python
# Use get_dummies() to create variables for text features
r = pd.get_dummies(mental_states_df, columns=[ 
'Frequency_Classical', 
'Frequency_Country', 
'Frequency_EDM', 
'Frequency_Folk', 
'Frequency_Gospel', 
'Frequency_Hiphop', 
'Frequency_Jazz', 
'Frequency_Kpop', 
'Frequency_Latin', 
'Frequency_Lofi', 
'Frequency_Metal', 
'Frequency_Pop', 
'Frequency_RB', 
'Frequency_Rap', 
'Frequency_Rock', 
'Frequency_Videogame',
'Anxiety',
'Depression',
'Insomnia'])
r
```

```python
# Standardize the data with StandardScaler()
mental_states_df_scaled = StandardScaler().fit_transform(r)
print(mental_states_df_scaled)
```

```python
# Reducing Data Dimensions Using PCA
# Using PCA to reduce dimension to two principal components
pca = PCA(n_components=3)
```

```python
pca_df = pca.fit_transform(mental_states_df_scaled)
pca_df
```

```python
# Create a DataFrame with the three principal components.
pcs_df = pd.DataFrame(
    data=pca_df, columns=["PC 1", "PC 2", "PC 3"])
pcs_df
```

```python
# Create an elbow curve to find the best value for K.
inertia = []
k = list(range(1, 11))

# Calculate the inertia for the range of K values
for i in k:
    km = KMeans(n_clusters=i, random_state=0)
    km.fit(pcs_df)
    inertia.append(km.inertia_)
```

```python
# Create the Elbow Curve using hvPlot
elbow_data = {"k": k, "inertia": inertia}
df_elbow = pd.DataFrame(elbow_data)
df_elbow.hvplot.line(x="k", y="inertia", xticks=k, title="Elbow Curve")
```

```python
# Initialize the K-Means model.
model = KMeans(n_clusters=4, random_state=0)

# Fit the model
model.fit(pcs_df)

# Predict clusters
predictions = model.predict(pcs_df)
predictions
```

```python
# Create a new DataFrame including predicted clusters and mental health states
# Concatentate the crypto_df and pcs_df DataFrames on the same columns.
clustered_df = pd.concat([mental_states_df, pcs_df], axis=1)

# Print the shape of the clustered_df
print(clustered_df.shape)
clustered_df.head(10)
```

```python
# Creating a 3D-Scatter with the PCA data and the clusters
scatter = px.scatter_3d(
    clustered_df,
    x="PC 1",
    y="PC 2",
    z="PC 3",
    width=800,
)
scatter.update_layout(legend=dict(x=0, y=1))
scatter.show()
```
